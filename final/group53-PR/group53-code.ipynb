**EDA Through T-tests**

In this section, we will perform a series of t-tests on the selected predictor variables: Race, Sex, Age, and Strip Search, along with the selected outcome variable: Severity of Crime.

This will help to determine which variables are statistically significant predictors of severity of crime, which will inform which tests to run next through ANCOVA.

***T-test 1: Race and Severity Of Crime.***


```python
# T-test: Race and Severity of Crime.

from scipy import stats

# Add a White versus BIPOC column to retain the original granularity of the data.
df2021_RankMF["White_BIPOC"] = df2021_RankMF["Perceived_Race"].map(lambda x:"BIPOC" if x != "White" else "White")

# Create data frames for White and BIPOC.
df2021_RankWh = df2021_RankMF[df2021_RankMF["White_BIPOC"] == "White"]
df2021_RankBI = df2021_RankMF[df2021_RankMF["White_BIPOC"] == "BIPOC"]

# Run Welch's T-test.
def welch_ttest(x, y): 
    dof = (x.var()/x.size + y.var()/y.size)**2 / ((x.var()/x.size)**2 / (x.size-1) + (y.var()/y.size)**2 / (y.size-1)) # Welch-Satterthwaite Degrees of Freedom.
    t, p = stats.ttest_ind(x, y, equal_var = False)

    print("\n",
          f"Welch's t-test= {t:.4f}", "\n",
          f"p-value = {p:.4f}", "\n",
          f"Welch-Satterthwaite Degrees of Freedom= {dof:.4f}")

welch_ttest(df2021_RankWh['Severity'], df2021_RankBI['Severity'])
```

```python
# Descriptive statistics for race and severity of crime.

# Mean and standard deviation.
print('Mean Severity Score for White and BIPOC :',df2021_RankWh['Severity'].mean(), df2021_RankBI['Severity'].mean())
print('Standard deviation Score for White and BIPOC:',df2021_RankWh['Severity'].std(), df2021_RankBI['Severity'].std())

# Confidence Interval.
import statsmodels.stats.weightstats as sms
cmWHBI = sms.CompareMeans(sms.DescrStatsW(df2021_RankWh['Severity']), sms.DescrStatsW(df2021_RankBI['Severity']))
print('Confidence interval for Severity Score:',cmWHBI.tconfint_diff(usevar='unequal'))
```

**Power Analysis**
---

In this section, we will conduct a power analysis on the group means of the severity score between those who were strip searched and those who were not.

```python
# Set up sample dataframes for records when a strip search is present and absent at the time of arrest.

SS_Yes = df2021_RankMF[df2021_RankMF["StripSearchYN"] == "Yes"]["Severity"]
SS_No = df2021_RankMF[df2021_RankMF["StripSearchYN"] == "No"]["Severity"]
```

```python
# Function to calculate the pooled standard deviation of 2 samples prior to calculating the Cohen's D.

def pooled_standard_deviation(sample1,sample2):
    #calculate the sample size
    n1, n2 = len(sample1), len(sample2)
    #calculate the variances
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)
    #calculate the pooled standard deviation
    numerator = ((n1-1) * var1) + ((n2-1) * var2)
    denominator = n1+n2-2
    return np.sqrt(numerator/denominator)

pooled_standard_deviation(SS_Yes, SS_No)
```

```python
# Function to calculate Cohen's D for independent samples.

def Cohens_d(sample1, sample2):
    u1, u2 = np.mean(sample1), np.mean(sample2)
    s_pooled = pooled_standard_deviation(sample1, sample2)
    #print(s_pooled)
    return ((u1 - u2) / s_pooled)

Effect_Size = Cohens_d(SS_Yes, SS_No)

# Calculating the Power with the effect size and sample size.

import statsmodels.stats.power as smp

power_analysis = smp.TTestIndPower()
power = power_analysis.solve_power(effect_size=(Cohens_d(SS_Yes, SS_No)), nobs1=len(df2021_RankMF), power=None, alpha=0.05)

print('Effect Size: {:.3f}'.format(Effect_Size))
print('Power: {:.3f}'.format(power))
```

```python
#Required sample size for each level in "Strip Search"

from statsmodels.stats.power import TTestIndPower
analysis = TTestIndPower()
effect_size = Cohens_d(SS_Yes, SS_No)
ratio_ = len(SS_No)/len(SS_Yes) #nob2/nob1

sample_size = analysis.solve_power(effect_size, power=0.8, nobs1=None, ratio=ratio_, alpha=0.05)
print('Sample Size of nobs1: %.3f' % sample_size, 'needed for SS_Yes') 
print('Actual size of SS_Yes:', len(SS_Yes))
nobs2 =ratio_*sample_size
print('Sample Size of nobs2: %.3f' % nobs2, 'needed for SS_No')
print('Actual size of SS_No:', len(SS_No))
```

```python
# Plot power curves.

sample_sizes = np.array(range(100,35000,5000))
effect_sizes = np.array([0.15,0.3,0.5,0.8])

fig14 = power_analysis.plot_power(dep_var='nobs', nobs=sample_sizes,
               effect_size=effect_sizes, alpha = 0.05, title = "Power Plot of Independent Sample T-test\n$\\alpha = 0.05$")


plt.show()
```

**Logistic Regression**

In this section, we will conduct a logistic regression to determine if demographic factors like race, sex, and age can predict the occurence of strip searches. 

***Recoding sex to a numerical variable.***

```python
# Ensure variables of interest (race, sex, age) are in numerical format.
# Race and age meet the requirements from the transformation steps prior to ANCOVA. We must recode sex to be numerical.

# Function to recode the categorical sex variable to a numerical variable.
def sex_recoded (Sex): 
    if Sex == "F": 
        return 1
    else:
        return 0

# Apply the function to recode the age group.
df2021_RankMF["Sex_Recoded"] = df2021_RankMF["Sex"].apply(sex_recoded)
df2021_RankMF.head() # Check new column.
```

***Set up and run the logistic regression.***

```python
from sklearn.model_selection import train_test_split

# Select the columns of the dataframe that we will use as predictors in our models.
x = df2021_RankMF[['AGRecoded', 'Race_Recoded', 'Sex_Recoded']].copy()

# Select the response column. 
y = df2021_RankMF.StripSearch

# Split the data into training and testing datasets.
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=123)
```

```python
# Building the model.

import statsmodels.formula.api as smf
import pandas as pd

# Create the formula string. 
all_columns = ' + '.join(x)
formula = "StripSearch ~ " + all_columns 
print("Formula: ", formula, "\n")

# Put the training predictors and responses into one dataframe to be input into the model.
trainingdata = pd.concat([x_train,y_train], axis = 1)

# Build the model.
log_reg_1 = smf.logit(formula, data=trainingdata).fit()
```

```python
# Plotting prediction interval for race.

%matplotlib inline
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

# Fit the model.
mod = sm.Logit.from_formula('StripSearch ~ Race_Recoded', data=trainingdata) 
res = mod.fit()
me = res.get_margeff()

Λ = lambda x: stats.logistic().cdf(x)
λ = lambda x: stats.logistic().pdf(x)
β_ = res.params.values.reshape(-1, 1)
V_ = res.cov_params().values

# For multiple observations.
def var_πs(xx, β, V_):
    α = λ(xx.dot(β))**2
    out = np.empty((500, 1))
    for i, x in enumerate(xx):
        out[i] = x.T.dot(V_).dot(x)
    return α * out

# Making some fake data.
xx = sm.add_constant(np.linspace(trainingdata.Race_Recoded.min(), trainingdata.Race_Recoded.max(), 500).reshape(-1, 1))
πs = Λ(xx.dot(β_))
vv = np.sqrt(var_πs(xx, β_, V_))

# Making the plot for race.
plt.plot(xx[:, 1], πs)
plt.fill_between(xx[:, 1], (πs - 1.96*vv).ravel(), (πs + 1.96*vv).ravel(),
                 alpha=.05, color='g')
```
