### Power Analysis

Prior to the ANCOVA, we have to do the power analysis. For our independent variable, there are 7 levels. 

```python
#code for Cohen's D 

#function to calculate the pooled standard deviation of 7 samples
def pooled_standard_deviation(sample1,sample2,sample3,sample4,sample5,sample6,sample7):
    #calculate the sample size
    n1, n2, n3, n4, n5, n6, n7 = len(sample1), len(sample2), len(sample3), len(sample4), len(sample5),len(sample6),len(sample7)
    #calculate the variances
    var1, var2, var3, var4, var5, var6, var7 = np.var(sample1, ddof=1), np.var(sample2, ddof=1), np.var(sample2, ddof=1), np.var(sample4, ddof=1), np.var(sample5, ddof=1), np.var(sample6, ddof=1), np.var(sample7, ddof=1)
    #calculate the pooled standard deviation
    numerator = ((n1-1) * var1) + ((n2-1) * var2) + ((n3-1) * var3) + ((n4-1) * var4) + ((n5-1) * var5) + ((n6-1) * var6) + ((n7-1) * var7)
    denominator = n1+n2+n3+n4+n5+n6+n7-7
    return np.sqrt(numerator/denominator)

# function to calculate Cohen's d for independent samples
def Cohens_d(sample1, sample2, sample3, sample4, sample5, sample6, sample7):
    u1, u2, u3, u4, u5, u6, u7 = np.mean(sample1), np.mean(sample2), np.mean(sample3), np.mean(sample4), np.mean(sample5), np.mean(sample6), np.mean(sample7)
    s_pooled = pooled_standard_deviation(sample1, sample2, sample3, sample4, sample5, sample6, sample7)
    #print(s_pooled)
    # calculate Cohen's d for each group
    d1 = (u1 - u7) / s_pooled
    d2 = (u2 - u7) / s_pooled
    d3 = (u3 - u7) / s_pooled
    d4 = (u4 - u7) / s_pooled
    d5 = (u5 - u7) / s_pooled
    d6 = (u6 - u7) / s_pooled
    return [d1, d2, d3, d4, d5, d6]
```

```python
# Subset the data for different age groups 
LT17 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 17 years and younger'])]
B1824 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 18 to 24 years'])]
B2534 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 25 to 34 years'])]
B3544 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 35 to 44 years'])]
B4554 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 45 to 54 years'])]
B5564 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 55 to 64 years'])]
GT65 = df_grouped[df_grouped['Age_group__at_arrest_'].isin(['Aged 65 years and older'])]
```

```python
#conduct power analysis 
from statsmodels.stats.power import TTestIndPower
effect_size = Cohens_d(LT17['StripSearchPercentage'],B1824['StripSearchPercentage'],B2534['StripSearchPercentage'],
                       B3544['StripSearchPercentage'],B4554['StripSearchPercentage'],B5564['StripSearchPercentage']
                       ,GT65['StripSearchPercentage'])
alpha = 0.05
power = 0.8
print('Effect size for each group comparison:',effect_size)
```

Prior to computing a ANCOVA to analyze whether the strip search percentage (outcome variable) differed between age groups (Seven-level explanatory variable), we calculated the effect size of the explanatory variable using Cohen's D metric. There are total 6 paired combinations of the levels. Level 7 which is aged over 65 is the reference group. All other groups will have effect size with the reference group. Effect size for people whose age less than 17 over whose age greater than 65 is 0.42. Effect size for people whose age less than 17 over whose age greater than 65 is 0.42. Effect size for people whose age between 18 and 24 over whose age greater than 65 is 0.3 ...

pg.homoscedasticity tests for homoscedasticity using Levene's test, which tests whether the variances of different groups are equal. In this case, it tests whether the variances of StripSearchPercentage are equal across different age groups (Age_group__at_arrest_). the p-value is less than 0.05, we can reject the null hypothesis that the variance is equal. So it is not homoscedasticity.

### Non-parametric test

If the assumptions of ANCOVA are not met, then non-parametric alternatives such as the Kruskal-Wallis test can be used. These tests do not assume normality or homoscedasticity, but they do have other assumptions and limitations that need to be considered. The Kruskal-Wallis test is used when there are more than two groups and the outcome variable is continuous. 

To apply the ranked Kruskal-Wallis, the values of variable 'Age_group__at_arrest_' will be convert to numeric. 

```python
from scipy.stats import rankdata, kruskal

# Rank the dependent variable and covariate
ranked_stripsearch = rankdata(df_grouped['StripSearchPercentage'])
ranked_female = rankdata(df_grouped['FemalePercentage'])
df_grouped['Age_group__at_arrest__numeric'] = df_grouped['Age_group__at_arrest_'].replace({'Aged 17 years and younger': 1, 'Aged 18 to 24 years': 2, 'Aged 25 to 34 years': 3, 
                      'Aged 35 to 44 years': 4, 'Aged 45 to 54 years':5, 'Aged 55 to 64 years':6,'Aged 65 years and older':7})
ranked_age = rankdata(df_grouped['Age_group__at_arrest__numeric'])
kruskal_result = kruskal(ranked_stripsearch, ranked_age, ranked_female)


print(kruskal_result)
```

The Kruskal-Wallis test with a ranked dependent variable and a ranked covariate as control variable resulted in a test statistic of 0.751 and a p-value of 0.687. This suggests that there is no significant difference in the StripSearchPercentage across different age groups at arrest when controlling for the effect of FemalePercentage.
